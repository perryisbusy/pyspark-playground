#### `dbutils.widgets.multiselect`

```
channel = ["Foo", "Bar", "Temp"]
dbutils.widgets.multiselect(name="channel", defaultValue="Temp", choices=[str(x) for x in channel] + ["None",  "All"])
```

```
dbutils.widgets.get(name="channel")
```
