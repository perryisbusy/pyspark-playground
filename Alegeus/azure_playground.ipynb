```python
%%time

from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

time = datetime.now()
date1 = time.strftime("%Y-%m-%d")+"-11"
# endpoint = "input/cobra/tpadata";
tpaFolder = 'A0C5284E-AC65-44A4-B1E5-39E5C1FD13D0'
theFileName = 'iris'

CONTAINER_NAME = "input/cobra/tpadata"
CONN_STRING='DefaultEndpointsProtocol=https;AccountName=algrptstdeabapov;AccountKey=LyUHbT7Jx5EoVEOwTGw1+lIhWjUmIKjXX5+hLz8tVxJalC2kpD+L7ccEyiCQZM/m6T9ihYrPrG73+AStWX8C8w==;EndpointSuffix=core.windows.net'

try:

    container_client = ContainerClient.from_connection_string(CONN_STRING, CONTAINER_NAME)
    output = iris_df.to_csv(index=False, encoding="utf-8-sig", sep='|', na_rep='NULL')
    blob_client = container_client.get_blob_client('{}/{}/{}.csv'.format(date1, tpaFolder, theFileName))

    blob_client.upload_blob(output)

except Exception as ex:
    
    print('Exception:')
    print(ex)
```

```python
def upload_df_into_blob(container_name, conn_string, df, sep):
    
    """
    upload a pandas dataframe to azure blob container
    """
    
    container_client = ContainerClient.from_connection_string(conn_str=conn_string, container_name=container_name)
    output = df.to_csv(index=False, encoding="utf-8-sig", sep=sep, na_rep='NULL')
    blob_client = container_client.get_blob_client('wine.csv')

    blob_client.upload_blob(output)
```

```python
def read_blob_into_df(container_name, conn_string, blob_location, sep):
    
    """
    read blob from azure and load as a pandas dataframe
    """
    
    from azure.storage.blob import BlobServiceClient, BlobClient, ContainerClient
    from io import StringIO
    import pandas as pd
    
    blob_service_client = BlobServiceClient.from_connection_string(conn_str=conn_string)
    blob_client = blob_service_client.get_blob_client(container=container_name, blob=blob_location)
    blob = blob_client.download_blob().content_as_bytes()
    blobstring = blob.decode("utf-8")
    df = pd.read_csv(filepath_or_buffer=StringIO(blobstring), sep=sep)
    
    return df
```
